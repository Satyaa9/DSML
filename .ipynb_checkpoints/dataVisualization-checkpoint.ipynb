```python
import numpy as np 
import pandas as pd 
import matplotlib.pyplot as plt 

price = [3000 , 4000 , 2000 , 6000]
year = [2015 , 2016 , 2017 , 2018]

plt.plot(year , price)
plt.show()
```

```python
batsman = pd.read_csv("sharma-kohli.csv")
batsman
```

```python
plt.plot(
    batsman["index"],
    batsman["V Kohli"],
    label="Virat Kohli"
)

plt.plot(
    batsman["index"],
    batsman["RG Sharma"],
    color="red",
    marker="o",
    linestyle="dashed",
    label="Rohit Sharma"
)

plt.title("Rohit Sharma and Virat Kohli IPL Stats")
plt.xlabel("Year")
plt.ylabel("Runs")

plt.legend()
plt.show()

```
