```python
import pandas as pd 
import numpy as np 
import seaborn as sns 
import matplotlib.pyplot as plt 
```

```python
data = pd.DataFrame({
    "Study Hours" : [1,2,3,4,5,6,7,8], 
    "Marks" : [32,38,45,52,60,70,45,60], 
    "Gender" : ["Male" , "Female" , "Male" , "Female","Male","Female","Male","Female"], 
    "Exam Type" : ["Online" , "Offline" , "Online" , "Offline","Online","Offline","Online","Offline"], 
    "Attendence" : [55, 60 , 65 , 70,60,76,80,90]
}); 
data
```

```python
sns.scatterplot(
    data=data , 
    x = "Study Hours" , 
    y = "Marks"
)
```

```python
sns.scatterplot(
    data=data , 
    x = "Study Hours" , 
    y = "Marks"
)
plt.title("Marks And Hours")
```

```python
sns.scatterplot(
    data=data , 
    x = "Study Hours" , 
    y = "Marks"
)
plt.title("Marks And Hours")
plt.xlabel("Study Hours")
plt.ylabel("Marks")
plt.show()
```

```python
sns.scatterplot(
    data=data , 
    x = "Study Hours" , 
    y = "Marks",
    hue = "Gender"
)
plt.title("Marks And Hours")
plt.xlabel("Study Hours")
plt.ylabel("Marks")
plt.show()
```

```python
sns.scatterplot(
    data=data , 
    x = "Study Hours" , 
    y = "Marks",
    hue = "Gender",
    style="Exam Type"
)
plt.title("Marks And Hours")
plt.xlabel("Study Hours")
plt.ylabel("Marks")
plt.show()
```

```python
sns.scatterplot(
    data=data , 
    x = "Study Hours" , 
    y = "Marks",
    hue = "Gender",
    style="Exam Type",
    size = "Marks"
)
plt.title("Marks And Hours")
plt.xlabel("Study Hours")
plt.ylabel("Marks")
plt.show()
```
