```python
import pandas as pd 
import numpy as np 
import seaborn as sns 
import matplotlib.pyplot as plt 
```

```python
data = {
    "Department" : ["IT","IT","IT","HR","HR","HR","SALES","SALES","SALES"],
    "Salary" : [50000,55000,60000,40000,42000,45000,30000,320000,35000]
}
df = pd.DataFrame(data)
print(df)
```

```python
sns.boxplot(
    data=df,
    x="Department",
    y="Salary",
    showmeans=True,
    meanprops={"marker": "o", "markerfacecolor": "red"}
     )
plt.title("Box Plot")
plt.xlabel("Department")
plt.ylabel("Salary")
plt.show()
```
